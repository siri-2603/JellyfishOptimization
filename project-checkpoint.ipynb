# Heart Disease Prediction using Artificial Jellyfish Search (JS) Optimization
This notebook implements the Artificial Jellyfish Search algorithm for feature selection on the Cleveland Heart Disease dataset. The objective is to maximize the fitness score while minimizing the number of features used.

```python
import numpy as np
import pandas as pd
import random

from sklearn.model_selection import StratifiedKFold
from sklearn.preprocessing import StandardScaler

from sklearn.svm import SVC
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import AdaBoostClassifier
from sklearn.neural_network import MLPClassifier

from sklearn.metrics import accuracy_score, confusion_matrix, roc_auc_score

from sklearn.exceptions import ConvergenceWarning
import warnings
warnings.filterwarnings("ignore", category=ConvergenceWarning)

# Global Seed for Reproducibility
SEED = 42
np.random.seed(SEED)
random.seed(SEED)
```

## 1. Data Loading and Preprocessing
Loading the Cleveland dataset, extracting the target variable, and applying standard scaling to the features.

```python
data = pd.read_csv("Heart_disease_cleveland_new.csv")
feature_names = list(data.drop("target", axis=1).columns)

data
```

```python
X = data.drop("target", axis=1).values
y = data["target"]

# Standardize the features
X = StandardScaler().fit_transform(X)
```

## 2. Helper Functions
Functions to evaluate the models using 5-Fold Stratified Cross-Validation and to initialize our base classifiers.

```python
def evaluate(model, X_data, y_data):
    skf = StratifiedKFold(n_splits=10, shuffle=True, random_state=SEED)
    acc_list, sen_list, spe_list, auc_list = [], [], [], []
    
    for tr, te in skf.split(X_data, y_data):
        model.fit(X_data[tr], y_data[tr])
        pred = model.predict(X_data[te])
        prob = model.predict_proba(X_data[te])[:, 1]
        
        tn, fp, fn, tp = confusion_matrix(y_data[te], pred).ravel()
        acc_list.append(accuracy_score(y_data[te], pred))
        sen_list.append(tp / (tp + fn))
        spe_list.append(tn / (tn + fp))
        auc_list.append(roc_auc_score(y_data[te], prob))
        
    return (np.mean(acc_list), np.mean(sen_list), np.mean(spe_list), np.mean(auc_list))

def get_models():
    return {
        "SVM":           SVC(probability=True, random_state=SEED),
        "Decision Tree": DecisionTreeClassifier(random_state=SEED),
        "AdaBoost":      AdaBoostClassifier(random_state=SEED),
        "ANN":           MLPClassifier(max_iter=500, random_state=SEED)
    }

# Initialize global dictionaries to store results across cells
baseline_results = {}
jfo_results = {}
```

## 3. Jellyfish Optimization Algorithm
Defining the metaheuristic optimizer. The fitness function balances 50% Accuracy and 50% AUC, applying a small penalty to penalize selecting too many features.

```python
class Jellyfish:
    def __init__(self, model, n_pop=10, max_iter=20, beta=3.0, gamma=0.1):
        self.model    = model
        self.n_pop    = n_pop
        self.max_iter = max_iter
        self.beta     = beta
        self.gamma    = gamma

    def fitness(self, sol):
        idx = np.where(sol > 0.5)[0]
        if len(idx) == 0:
            return 0.0
            
        X_sel = X[:, idx]
        skf   = StratifiedKFold(n_splits=10, shuffle=True, random_state=SEED)
        accs, aucs = [], []
        
        for tr, te in skf.split(X_sel, y):
            m = type(self.model)(**self.model.get_params())
            m.fit(X_sel[tr], y[tr])
            pred = m.predict(X_sel[te])
            prob = m.predict_proba(X_sel[te])[:, 1]
            accs.append(accuracy_score(y[te], pred))
            aucs.append(roc_auc_score(y[te], prob))

        # combined: 50% accuracy + 50% AUC
        combined = 0.5 * np.mean(accs) + 0.5 * np.mean(aucs)
        return combined - 0.01 * (len(idx) / X.shape[1])

    def optimize(self, seed):
        np.random.seed(seed)
        random.seed(seed)
        dim    = X.shape[1]
        pop    = np.random.rand(self.n_pop, dim)
        scores = [self.fitness(pop[i]) for i in range(self.n_pop)]

        best_idx   = int(np.argmax(scores))
        best       = pop[best_idx].copy()
        best_score = scores[best_idx]

        print(f"  {'Iter':<6} {'Best Score':>11}  {'Features':>10}")
        print(f"  {'-'*32}")

        for t in range(self.max_iter):
            mean = np.mean(pop, axis=0)
            for i in range(self.n_pop):
                ct = abs((1 - t / self.max_iter) * (2 * random.random() - 1))

                if ct >= 0.5:
                    trend  = best - self.beta * random.random() * mean
                    pop[i] = pop[i] + random.random() * trend
                else:
                    if random.random() < 0.5:
                        pop[i] = pop[i] + self.gamma * random.random() * 1.0
                    else:
                        j = random.randint(0, self.n_pop - 1)
                        if scores[i] < scores[j]:
                            pop[i] += random.random() * (pop[j] - pop[i])
                        else:
                            pop[i] += random.random() * (pop[i] - pop[j])

                pop[i]    = np.clip(pop[i], 0, 1)
                sc        = self.fitness(pop[i])
                scores[i] = sc
                
                if sc > best_score:
                    best_score = sc
                    best       = pop[i].copy()

            n_feat = int((best > 0.5).sum())
            print(f"  {t+1:<6} {best_score:>11.4f}  {n_feat:>4} / {dim}")

        return best, best_score
```

## 4. Execution Step 1: Baseline Models
Testing the base classifiers using all 13 original features to establish a benchmark.

```python
print()
print("=" * 65)
print("  STEP 1 — WITHOUT JFO  |  All features  |  Default params")
print("=" * 65)
print(f"  {'Model':<16} {'Acc':>7}  {'Sen':>7}  {'Spe':>7}  {'AUC':>7}")
print("-" * 65)

for name, model in get_models().items():
    acc, sen, spe, auc = evaluate(model, X, y)
    baseline_results[name] = (acc, sen, spe, auc)
    print(f"  {name:<16} {acc*100:>6.2f}%  {sen*100:>6.2f}%  "
          f"{spe*100:>6.2f}%  {auc*100:>6.2f}%")
print("=" * 65)
```

## 5. JFO Model Summary Function
Defining a reusable function to run JFO and print the summary for any specified model.

```python
def run_and_print_jfo(model_name, seed):
    models = get_models()
    model = models[model_name]
    
    print()
    print("=" * 65)
    print(f"  STARTING JFO  |  Model: {model_name}")
    print("=" * 65)

    jf     = Jellyfish(model=model)
    sol, _ = jf.optimize(seed=seed)

    idx       = np.where(sol > 0.5)[0]
    sel_names = [feature_names[i] for i in idx]

    m = type(model)(**model.get_params())
    acc, sen, spe, auc = evaluate(m, X[:, idx], y)
    jfo_results[model_name]  = (acc, sen, spe, auc, idx)

    b_acc, b_sen, b_spe, b_auc = baseline_results[model_name]
    
    d_acc = (acc - b_acc) * 100
    d_sen = (sen - b_sen) * 100
    d_spe = (spe - b_spe) * 100
    d_auc = (auc - b_auc) * 100

    print()
    print("=" * 65)
    print(f"  JFO RESULTS  |  {model_name}")
    print("=" * 65)
    print(f"  {'Phase':<16} {'Acc':>7}  {'Sen':>7}  {'Spe':>7}  {'AUC':>7}")
    print("-" * 65)
    print(f"  {'Baseline':<16} {b_acc*100:>6.2f}%  {b_sen*100:>6.2f}%  "
          f"{b_spe*100:>6.2f}%  {b_auc*100:>6.2f}%")
    print(f"  {'With JFO':<16} {acc*100:>6.2f}%  {sen*100:>6.2f}%  "
          f"{spe*100:>6.2f}%  {auc*100:>6.2f}%")
    print(f"  {'Delta':<16} {d_acc:>+6.2f}%  {d_sen:>+6.2f}%  "
          f"{d_spe:>+6.2f}%  {d_auc:>+6.2f}%")
    print("-" * 65)
    print(f"  Features selected : {len(idx)}/13")
    print(f"  {sel_names}")
    print("=" * 65)
    print()
```

## 6. Model Optimizations
Run the JFO feature selection for each model individually.

```python
# Run Optimization for SVM
run_and_print_jfo("SVM", seed=1)
```

```python
# Run Optimization for Decision Tree
run_and_print_jfo("Decision Tree", seed=1)
```

```python
# Run Optimization for AdaBoost
run_and_print_jfo("AdaBoost", seed=0)
```

```python
# Run Optimization for ANN
run_and_print_jfo("ANN", seed=0)
```
